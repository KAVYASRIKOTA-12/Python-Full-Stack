```python
import matplotlib.pyplot as plt
```

```python
x=[2,4,6,8,9]
y=[3,4,5,6,7]
y1=[6,7,8,9,10]

plt.plot(x,y)
plt.show()
```

```python
plt.subplot(2,3,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y1)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.suptitle("Data")
plt.title("Graph 1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)

plt.subplot(2,2,2)
plt.title("Graph 2")
plt.plot(x,y1)
plt.show()
```
